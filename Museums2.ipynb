### Clean dataframes

```python
museums_cols = ["Museum Name", "Museum Type", "City (Administrative Location)",
                "State (Administrative Location)", "Income", "Revenue"]
museums_cleaned = museums_df[museums_cols].copy()

# Rename the column headers
museums_cleaned = museums_cleaned.rename(columns={"Museum Name": "Museum Name",
                                                  "Museum Type": "Museum Type",
                                                  "City (Administrative Location)": "City",
                                                  "State (Administrative Location)": "State",
                                                  "Income": "Income",
                                                  "Revenue": "Revenue"})

# Clean the data by dropping duplicates and setting the index
museums_cleaned.drop_duplicates("Museum Name", inplace=True)
#museums_cleaned.set_index("Museum Name", inplace=True)

museums_cleaned.head()
```

```python
#museums_cleaned['id'] = museums_cleaned.index
#museums_cleaned.head()
```

```python
museums_cleaned.dtypes
```

```python
museums_cleaned.count()
```

```python
tripadvisor_cols = ["MuseumName", "Rating", "ReviewCount"]
tripadvisor_cleaned = tripadvisor_df[tripadvisor_cols].copy()

# Rename the column headers
tripadvisor_cleaned = tripadvisor_cleaned.rename(columns={"MuseumName": "Museum Name",
                                                          "Rating": "Rating",
                                                          "ReviewCount": "Review Count"})

# Clean the data by dropping duplicates and setting the index
tripadvisor_cleaned.drop_duplicates("Museum Name", inplace=True)
#tripadvisor_cleaned.set_index("Museum Name", inplace=True)

tripadvisor_cleaned['Museum Name'] = tripadvisor_cleaned['Museum Name'].str.upper()

tripadvisor_cleaned.head()
```

```python
#tripadvisor_cleaned['id'] = tripadvisor_cleaned.index
#tripadvisor_cleaned.head()
```

```python
tripadvisor_cleaned.count()
```

### Create database connection

```python
connection_string = "postgres:postgres@localhost:5432/museums_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

### Load DataFrames into database

```python
museums_cleaned.to_sql(name='museums_df', con=engine, if_exists='append', index=False)
```

```python
tripadvisor_cleaned.to_sql(name='tripadvisor_df', con=engine, if_exists='append', index=False)
```

### Check tables from postgres

```python
pd.read_sql_query('select * from museums_df', con=engine).head()
```
